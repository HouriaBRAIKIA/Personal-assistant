```python
num = 1

#créer une fenetre
fenetre = Toplevel()

#Interface
def interface():

    #personnalisé la fenetre
    fenetre.minsize(390, 450)
    fenetre.resizable(width=False, height=False)
    fenetre.config(background = '#0f056b')
    screen_w = fenetre.winfo_screenwidth()
    screen_h = fenetre.winfo_screenheight()
    posX = (screen_w / 2) - (390 / 2)
    posY = (screen_h / 2) - (450 / 2)
    fenetre.geometry('%dx%d+%d+%d' % (390, 450, posX, posY))

    #Button
    btn_img = PhotoImage(file="btn.png")
    ouvrir_btn = Label(fenetre, image=btn_img, bd=0, cursor="hand2")
    ouvrir_btn.bind("<Button-1>", lambda event: cerveau())
    ouvrir_btn.place(x=100, y=280)

    logo = PhotoImage(file="logo.png")
    Label(fenetre, image=logo, relief='flat', bd=0).place(x=53, y=20)
    
    assistant_speaks("Hello Houria.")
    fenetre.iconbitmap("icone.ico")
    fenetre.title("Personnel assistante")

    #afficher la fenetre
    fenetre.mainloop()
```

```python
#Cerveau
def cerveau():
    try :
        assistant_speaks("What can i do for you ?")
        text = entendre()

        if "bye" in str(text) or "tchaw " in str(text) or "exit" in str(text):
            assistant_speaks("Okay bye, Houria.")
            quitter()
        elif "search" in text.lower() or "what is" in text.lower() :
            google(text)
        elif "open" in text.lower() :
            ouv_application(text)
        elif "close" in text.lower() :
            ferm_application(text)
        elif "who are you" in text.lower() or "introduce your self" in text.lower() :
            assistant_speaks("Je suis l\'assistante personnelle de Houria , est que je peux vous aider ?")
            cerveau()
        elif "thanks" in text.lower() or "thank you" in text.lower() :
            assistant_speaks("Il n'y a pas de quoi")
        elif "who made you" in text.lower() :
            assistant_speaks("I was built by Houria")
        elif "It's okay" in text.lower() :
            assistant_speaks("Good hamdoulilah, and you ?")
            text = entendre()
            if "okay" in text.lower() or "good" in text.lower() :
                assistant_speaks("Hamdoulilah, what can i do for you ?")
                cerveau()
            else : 
                assistant_speaks("Why, what can i do for you ?")
                cerveau()
        elif "whrite" in text.lower():
            ecrire()
        elif "read" in text.lower():
            lire() 
        elif "email" in text.lower() :
            email(text)
        elif "photo" in text.lower() :
            photo()
        elif "screenshot" in text.lower() :
            capture()
        elif "battery" in text.lower():
            info_batt()
        elif "find" in text.lower():
            find_in_pc()
        elif "date" in text.lower() or "time" in text.lower() or "day" in text.lower():
            date_heure()
        elif text != 0 : 
            assistant_speaks(text)
        
    except : 
        assistant_speaks("Try again")
```

```python
#Quitter
def quitter():
    fenetre.destroy()
    os.system("^C")

#Conversion_Voix->Texte
def entendre():
    r = sr.Recognizer()
    with sr.Microphone() as source:                                                                                                 
        audio = r.listen(source)
    try:
        return "You said " + r.recognize_google(audio, language='en-US')
    except :
        assistant_speaks("Sorry i can't understand")
        cerveau()
        return 0

#Conversion_Text->Voix
def assistant_speaks(output):
    global num
    num+=1
    toSpeak = gTTS(text=output, lang='en-US', slow=False)
    file = "file"+str(num)+".mp3"
    toSpeak.save(file)
    playsound(file, True)
    os.remove(file)
```

```python
#Chercher_dans_Google
def google(cmd):
    try:
        if 'YouTube' in cmd :
            if "vidéo" in cmd :
                assistant_speaks("Opening youtube")
                indx = cmd.lower().split().index('youtube')
                cherche = cmd.split()[indx+1:]
                webbrowser.open_new("https://www.youtube.com/results?search_query=" + '+'.join(cherche))
        elif 'wikipédia' in cmd.lower() :
            assistant_speaks("Opening Wikipedia")
            indx = cmd.lower().split().index('wikipédia')
            cherche = cmd.split()[indx+1:]
            webbrowser.open_new("https://fr.wikipedia.org/wiki/" + '_'.join(cherche))
        elif 'instagram' in cmd.lower() :
            assistant_speaks("Opening instagram")
            webbrowser.open_new("https://www.instagram.com")
        elif 'pinterest' in cmd.lower() :
            webbrowser.open_new("https://www.pinterest.fr")
        else :
            assistant_speaks("Opening Google")
            indx = cmd.lower().split().index('google')
            cherche = cmd.split()[indx+1:]
            webbrowser.open_new("https://www.google.com/search?client=opera&q="+ '+'.join(cherche) +"&sourceid=opera&ie=UTF-8&oe=UTF-8")
    except: assistant_speaks("I can't search for you")
```

```python
#Prendre une photo
def photo():
    assistant_speaks("Prepare yourself then click on the letter Q")
    try:
        videoCaptureObject = cv2.cv2.VideoCapture(0, cv2.cv2.CAP_DSHOW)
        result = True
        while(result):
            ret,frame = videoCaptureObject.read()
            cv2.cv2.imshow('Capturing Video',frame)
            if(cv2.cv2.waitKey(1) & 0xFF == ord('q')):
                cv2.cv2.imwrite("NewPicture.jpg",frame)
                break
        videoCaptureObject.release()
        cv2.cv2.destroyAllWindows()
    except : print("pb")

def capture():
    assistant_speaks("Take a screenshot")
    x =ImageGrab.grab()
    assistant_speaks("What name do you want to name this image?")
    txt = entendre()
    nom_img = txt.split()[3:]
    nom = " ".join(nom_img) + ".png"
    x.save(nom)
    x.show()
```

```python
#information battrie
def info_batt():
    bat = psutil.sensors_battery()
    percent = str(bat.percent)
    assistant_speaks("The battery level is " + percent + "%")

#Date et heure
def date_heure():
    assistant_speaks("Today is " + datetime.now().strftime("%d %B, %Y")) 
    assistant_speaks("The current time is " + datetime.now().strftime("%H:%M:%S"))

#Arreter l ordi
def arreter():
    assistant_speaks("Are you sure you want to shut down the computer?")
    repns : str = entendre().lower()
    if "yes" in str(repns) :
        os.system("shutdown /s /t 1")
    else :
        return
```

```python
#Lire un text
def lire():
    assistant_speaks("Please choose a file")
    nom_fichier = askopenfilename(title="Please choose a file", filetypes=[("Document texte", ".txt"), ("Document Microsoft Office Word", ".docx")])
    if len(nom_fichier) > 0:
        texte = open(nom_fichier, "r")
        assistant_speaks(texte.read())
        texte.close()

#Ecrire dans un fichier
def ecrire():
    assistant_speaks("Please choose a file")
    nom_fichier = askopenfilename(title="Please choose a file", filetypes=[("Document texte", ".txt")])
    if len(nom_fichier) > 0:
        fichier = open(nom_fichier, "a")
        assistant_speaks("Dictate me your text")
        texte = entendre()
        fichier.write("\n" + texte)
        assistant_speaks("Finish")
        fichier.close()
```
